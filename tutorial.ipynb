# Multiple dimensional data

When dealing with high-dimensional datasets, keplergl has a few features we can use to create high-information graphics (Tufte, Chapter 8).

One is the color which you can use to make a distinction between the starting and ending points of the arcs in the diagram. Another thing we can do is change the thickness of the strokes based on a variable. Here I will show how to change the thickness based on the number of times a trip has been taken.

First I start by creating a new version of the dataframe that combines each identical trip and adds a column of the number of times the trip has been taken.

```python
df_agg = df.groupby([
    'start_station_name', 'end_station_name',
    'start_lat', 'start_lng',
    'end_lat', 'end_lng'
]).size().reset_index(name='trip_count')
df_agg.head()
```

Then I create a new map object using this new dataframe.

```python
map_2 = keplergl.KeplerGl(height=600, data={'citibike': df_agg})
map_2
```

To change the thickness strokes by the number of trips you go to your arc layer and click on the three dots by "Stroke" and go to the "Stroke Based On" dropdown menu. Here you can select the "trip_count" as the variable the stroke is based on. You can also change the range of stroke sizes and the scale (linear, log, or square root) it uses.

# Saving your Connection Map

Each time you run the cell containing your map object it will reset any changes you have made to the layers or filtering. To counteract this and make sure your map loads the same each time you have to add a configuration. After you have chosen your layers and filtering, if you run `your_map.config` you will see the current configuration of your map which includes any layers you have created and whether or not they are visible on the map, as well as any filters you have selected.

```python
map_2.config
```

Then, if you specify this configuration in the `keplergl.KeplerGl()` object, your map will initiate with the specified layers and filters.

```python
map_2 = keplergl.KeplerGl(height=600, data={'citibike': df_agg}, config=map_2.config)
map_2
```

If you then want to save this configured map as an html file you can use the `save_to_html` method.

```python
map_1.save_to_html(file_name='map_1.html', config=map_1.config)
map_2.save_to_html(file_name='map_2.html', config=map_2.config)
```
