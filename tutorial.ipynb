```python
import pandas as pd
import keplergl
```

```python
df = pd.read_csv('data/202603-citibike-tripdata_1.csv')
```

```python
map_1 = keplergl.KeplerGl(height=600, data={'citibike': df})
map_1
```

```python
df_agg = df.groupby([
    'start_station_name', 'end_station_name',
    'start_lat', 'start_lng',
    'end_lat', 'end_lng'
]).size().reset_index(name='trip_count')
```

```python
map_2 = keplergl.KeplerGl(height=600, data={'citibike': df_agg})
map_2
```

```python
map_2.config
```

```python
map_2 = keplergl.KeplerGl(height=600, data={'citibike': df_agg}, config=map_2.config)
```

```python
map_2.save_to_html(file_name='citibike_map.html')
```
